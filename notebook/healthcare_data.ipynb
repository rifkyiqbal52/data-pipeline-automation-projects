```python
from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'csv-healthcare_dataset'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'healthcare'
path_to_data = '/content/P2M3_Rifky-Iqbal_data_raw_cleaning.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-healthcare'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

```python
# Expectation 1 : to be Unique

validator.expect_column_values_to_be_unique('visit_id')
```

```python
# Expectation 2 : to be between

validator.expect_column_values_to_be_between(
    column='Age', min_value=1, max_value=90
)
```

```python
# Expectation 3 : to be in set

validator.expect_column_values_to_be_in_set('Gender', ['Male', 'Female'])
```

```python
# Expectation 4 : to be type list

validator.expect_column_values_to_be_in_type_list('Billing Amount', ['integer', 'float'])
```

```python
# Expectation 5 : to not be null

validator.expect_column_values_to_not_be_null('Name')
```

```python
# Expectation 6 : to be match regex

validator.expect_column_values_to_match_regex(
    column="Medication",
    regex=r"^[A-Z][a-z]+(?: [A-Z][a-z]+)*$")
```

```python
# Expectation 7 : to be max to be  between

validator.expect_column_max_to_be_between("Room Number", min_value=1, max_value=500)
```
